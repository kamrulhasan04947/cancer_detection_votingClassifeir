```python
import glob
from sklearn.preprocessing import LabelEncoder
```

```python
SIZE = 224
train_images = []
train_labels = []
for directory_path in glob.glob("/content/drive/MyDrive/Data sets/dataset-20230810T102546Z-001/dataset/train_data/first/*"):
    label = directory_path.split("\\")[-1]
    for img_path in glob.glob(os.path.join(directory_path)):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        cv2_imshow(img)
        train_images.append(img)
        train_labels.append(label)
```

```python
SIZE = 224
test_images = []
test_labels = []
for directory_path in glob.glob("/content/drive/MyDrive/Data sets/dataset-20230810T102546Z-001/dataset/train_data/second/*"):
    test_label = directory_path.split("\\")[-1]
    for img_path in glob.glob(os.path.join(directory_path)):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        cv2_imshow(img)
        test_images.append(img)
        test_labels.append(test_label)
```

```python
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
train_path ="/content/drive/MyDrive/Data sets/dataset-20230810T102546Z-001/dataset/train_data/"
test_path ="/content/drive/MyDrive/Data sets/dataset-20230810T102546Z-001/dataset/train_data/"
```

```python
train_set = train_datagen.flow_from_directory(train_path,
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = train_datagen.flow_from_directory(test_path,
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
feature_extractor=VGG_model.predict(train_set)
```

```python
VGG_model.add(tf.keras.layers.Dense(10, activation = 'softmax'))
```

```python
r = VGG_model.fit(
  train_set,
  validation_data=test_set,
  epochs=5,
  steps_per_epoch=len(train_set),
  validation_steps=len(test_set)
)
```
